### 读取商品-城市信息

```python
#由于数据太大，这里抽样读取一部分：
# 如果数据量大于10万条，则进行随机抽样
import pandas as pd

# 文件路径，根据实际情况修改
file_path = '零售数据匹配对应商店城市信息.csv'

# 设定分块大小，可以根据实际内存情况进行调整
chunksize = 10000
chunks = pd.read_csv(file_path, chunksize=chunksize,index_col=[0])

sampled_data = None
total_rows = 0

# 遍历每个数据块，并计算总行数
for chunk in chunks:
    total_rows += len(chunk)

# 计算抽样的比例
sample_ratio = 100000 / total_rows

# 重新读取数据，进行分块抽样
chunks = pd.read_csv(file_path, chunksize=chunksize)
sampled_chunks = []

for chunk in chunks:
    sampled_chunk = chunk.sample(frac=sample_ratio, random_state=42)
    sampled_chunks.append(sampled_chunk)

# 将抽样后的数据块合并为一个DataFrame
sampled_data = pd.concat(sampled_chunks, ignore_index=True)

# 对抽样后的数据进行后续操作
print(sampled_data.head())
```

```python
sampled_data.head()
```

```python
#观察数据，零售里面staff_id有空值，这里要去掉
sampled_data[sampled_data["staff_id"].isna()==True].head()
```

```python
#剔除空值
sampled_data=sampled_data[sampled_data["staff_id"].isna()!=True]
```

```python
#因为要匹配商品，所以staff id 需要转成int或者str型
sampled_data["staff_id"]=sampled_data["staff_id"].astype("int")
```

```python
#对时间要改写成年一列，月一列
df1['billdate']=pd.to_datetime(df1['billdate'])
df1['year'] = df1['billdate'].dt.year
df1['month'] = df1['billdate'].dt.month
```

```python
df_groupby=df1.groupby(["month","store_id","staff_id"]).agg({"retail_count":np.sum,"adcode":np.max,"province":np.max,"city":np.max})
```

```python
#按商店id，月份，staff_id聚合，然后再按prod_type聚合
df_groupby=df_groupby.reset_index()
```

```python
# 计算每列缺失值的比例
missing_ratio = df3.isnull().sum() / len(df3)

# 筛选出缺失值大于70%的列
columns_to_drop = missing_ratio[missing_ratio > 0.8].index

# 删除缺失值大于70%的列
df3.drop(columns_to_drop, axis=1, inplace=True)
```
